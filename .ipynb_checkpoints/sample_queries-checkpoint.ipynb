```python
%sql SELECT * FROM songplays;
```

## IDENTIFYING USERS WITH THE MOST RECORDS

```python
%%sql SELECT user_id, 
COUNT(user_id) AS total_user
FROM songplays
GROUP BY user_id
ORDER BY total_user DESC
LIMIT 5;
```

## IDENTIFYING THE TOP 5 USER LOCATIONS

```python
%%sql SELECT location,
COUNT(location) AS total_number
FROM songplays  
GROUP BY location 
ORDER BY total_number DESC
LIMIT 5;
```

## IDENTIFYING THE TOP 5 WEB BROWSERS

```python
%%sql SELECT user_agent,
COUNT(user_agent) AS total_agent
FROM songplays  
GROUP BY user_agent 
ORDER BY total_agent DESC
LIMIT 5;
```

## IDENTIFYING NUMBER USERS WITH PAID AND FREE SUSCRIPTIONS 

```python
%%sql SELECT level,
COUNT(level) AS total_level
FROM songplays  
GROUP BY level 
ORDER BY total_level DESC
LIMIT 5;
```

## IDENTIFYING THE GENDER OF USERS WITH TOP MOST RECORDS 

```python
%%sql SELECT 
users.gender,
COUNT(users.gender) AS total
FROM songplays  
INNER JOIN users
ON songplays.user_id = users.user_id
GROUP BY users.gender
ORDER BY total DESC
LIMIT 5;
```

```python
%%sql SELECT 
users.gender,
songplays.user_id,
COUNT(songplays.user_id) AS total_user
FROM songplays  
INNER JOIN users
ON songplays.user_id = users.user_id
GROUP BY songplays.user_id
ORDER BY total_user DESC
LIMIT 5;
```
